```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Exchange.csv')
```

```python
a=data.groupby('Exchange_name').sum()
```

```python
a=a.reset_index()
```

```python
a=a.sort_values(by=['# Coins'],ascending=False)
```

```python
#Top 10 exchanges with maximum number of currencies
plt.figure(figsize=(20,6))
sns.barplot(x="Exchange_name", y="# Coins", data=a[:10])

plt.show()
```

```python
# Country wise count of Exchanges 
a=data.groupby('Country').count()
a=a.reset_index()
a=a.sort_values(by=['Exchange_name'],ascending=False)
a=a.rename(columns={"Exchange_name":"No_of_Exchanges"})
plt.figure(figsize=(20,6))
sns.barplot(x="Country", y="No_of_Exchanges", data=a)

plt.show()
```

```python
data_2=pd.read_csv('complete_exchange.csv')
```

```python
data_2=data_2.groupby('Centralised_Decentralised').count()
```

```python
data_2=data_2.reset_index()
```

```python
data_2
```

```python
# % of centralised and decentralised exchanges
labels = data_2['Centralised_Decentralised']
sizes = data_2['index']
# explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
