```python
#packages required
import requests
from bs4 import BeautifulSoup
import pandas as pd
import urllib
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import os
import re
```

```python
# main url where all the exchanges are listed
url="https://www.coingecko.com/en/exchanges"
```

```python
#Requirement : Chrome driver should be installed in the same directory in which code is running
driver = webdriver.Chrome() #This code will show error if chrome driver is not installed
driver.get(url)
```

```python
#to extract links of all the exchanges
soup_level1=BeautifulSoup(driver.page_source, 'lxml')
links=set()
for link in soup_level1.findAll('a',attrs={'href': re.compile("/exchanges/")}):
    links.add('https://www.coingecko.com'+link.get('href'))
```

```python
#to scrape data from all the exchanges individually
req=requests.get(list(links)[0])
data=pd.read_html(req.text)
main_table=pd.DataFrame(data[0]['Coin'])
main_table['Exchange']=list(links)[0].split('/')[-1]
# final_data=[]
# for link in (list(links)[1:]):
#     req=requests.get(link)
#     data=pd.read_html(req.text)
#     if (len(data[0])>0):
#         table=pd.DataFrame(data[0][data[0].columns[0]],columns=['Coin'])
#         table['Exchange']=link.split('/')[-1]
#         final_data.append(table)
```

```python
for link in (list(links)[1:]):
    req=requests.get(link)
    data=pd.read_html(req.text)[0]
    if (len(data)>0):
        table=data[data.columns[0]]
        table=pd.DataFrame(table)
        table['Exchange']=link.split('/')[-1]
        main_table=pd.concat([main_table,table])
```
